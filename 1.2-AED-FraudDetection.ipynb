```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("https://media.githubusercontent.com/media/diegonogare/MachineLearning/main/creditcard.csv")
```

```python
df.info()
df.describe()
df.head()
```

```python
corr = df.corr()
corr.style.background_gradient()
```

```python
df2 = df[['V3','V4','V7','V10','V11','V12','V14','V16','V17','V18','Class']]
corr = df2.corr()
corr.style.background_gradient()
```

```python
X = df2.drop(['Class'], axis = 1)
Y = df2[['Class']]

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
X_train.describe()
```

```python
y_train.describe()
```

```python
X_test.describe()
```

```python
y_test.describe()
```

```python
for i in X:
  plt.scatter(df2[i],df2['Class'])
  plt.title('Correlação entre ' + i +' e Class')
  plt.xlabel(i)
  plt.ylabel('Class')
  plt.show()
```

```python
legenda = ["Média","Mediana"]

for i in X:
  plt.hist(df2[i])
  plt.axvline(df2[i].mean(), color='purple', lw=3, ls="--") #Média
  plt.axvline(df2[i].median(), color='orange', lw=3, ls="--") #Mediana
  plt.legend(legenda, loc="upper right")
  plt.title('Histograma de ' + i)  
  plt.show()
```
